## 任务 1：PromptTemplate 实例创建（2 种方式）

要求：分别使用 PromptTemplate() 配合 .format() 初始化、PromptTemplate.from_template() 两种方式创建「美食推荐提示词模板」，模板需包含 city（城市）、food_type（美食类型）两个变量，模板内容为："请为{city}推荐3道经典的{food_type}美食，并简要说明推荐理由"。

要求 1：两种方式创建的模板实例，变量列表需一致；

要求 2：打印两种方式创建的模板对象及模板内容，验证一致性。


```python
# 方式1：直接使用format()方法
from langchain_core.prompts import PromptTemplate
template1 = PromptTemplate(
    template="请为{city}推荐3道经典的{food_type}美食，并简要说明推荐理由。",
    input_variables=["city", "food_type"]
)
prompt = template1.format(city="北京", food_type="川菜")
print(prompt)
print(type(prompt))
```

```python
# 方式2：使用invoke()方法
from langchain_core.prompts import PromptTemplate
template2 = PromptTemplate(
    template="请为{city}推荐3道经典的{food_type}美食，并简要说明推荐理由。",
    input_variables=["city", "food_type"]
)   

prompt = template2.invoke({"city":"北京", "food_type":"川菜"})
print(prompt)
print(type(prompt))
```

## 任务 2：特殊提示词模板使用（部分变量赋值 + 组合模板）

要求：

2.1 部分提示词模板使用

基于任务 1 的模板，先为 food_type 赋值为「家常菜」，生成仅包含 city 变量的新模板（即固定美食类型，仅留城市变量）。

2.2 组合提示词模板使用

创建一个「补充说明模板」："\n补充要求：推荐的美食需符合{feature}（如：性价比高、适合聚餐），且标注人均消费"，将其与 2.1 生成的新模板组合为一个完整模板（需包含 city 和 feature 两个变量）。
最终完整模板示例效果："请为{city}推荐3道经典的家常菜美食，并简要说明推荐理由\n补充要求：推荐的美食需符合{feature}（如：性价比高、适合聚餐），且标注人均消费"。

```python
# 方式：使用partial_variables参数预填充部分变量
from langchain_core.prompts import PromptTemplate
template2 = PromptTemplate(
    template="请为{city}推荐3道经典的{food_type}美食，并简要说明推荐理由。",
    partial_variables={"food_type": "川菜"}#将部分变量初始化
)   

prompt = template2.invoke({"city":"北京"})
print(prompt)
print(type(prompt))
```

```python
# 方式：使用.partial()方法预填充部分变量
from langchain_core.prompts import PromptTemplate
template2 = PromptTemplate(
    template="请为{city}推荐3道经典的{food_type}美食，并简要说明推荐理由。",).partial(food_type="川菜" ) #直接在实例后面用.partial()方法预
prompt = template2.invoke({"city":"北京"})
print(prompt)
print(type(prompt))
```

```python
from langchain_core.prompts import PromptTemplate
template = (
    PromptTemplate.from_template("请为{city}推荐3道经典的{food_type}美食，并简要说明推荐理由。").partial(food_type="川菜")
    + "\n\n补充要求：推荐的美食需符合{feature}（如：性价比高、适合聚餐），且标注人均消费。")
prompt = template.invoke({"city":"北京", "feature":"性价比高"})
print(prompt)
print(prompt.to_string())
print(type(prompt))
```

```python
# 添加异常处理

import os
import dotenv   
from langchain_openai import ChatOpenAI
from langchain_core.prompts import PromptTemplate

dotenv.load_dotenv(os.path.expanduser("~/.env"))

os.environ["OPENAI_API_KEY"] = os.getenv("OPENAI_API_KEY")
os.environ["OPENAI_BASE_URL"] = os.getenv("OPENAI_BASE_URL")

llm = ChatOpenAI(model_name="gpt-4o-mini")

template = (
    PromptTemplate.from_template("请为{city}推荐3道经典的{food_type}美食，并简要说明推荐理由。").partial(food_type="川菜")
    + "\n\n补充要求：推荐的美食需符合{feature}（如：性价比高、适合聚餐），且标注人均消费。")
prompt = template.invoke({"city":"北京", "feature":"性价比高"})
print(prompt.to_string())

def call_llm(prompt):
    try:
        response = llm.invoke(prompt)
        return response.content
    except Exception as e:
        print(f"调用语言模型时出错: {e}")
        return None

call_llm(prompt)
```
